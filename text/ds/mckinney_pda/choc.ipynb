```python
def chocolateBars(bars, wrappers, total):
    if bars == 0:
        # print(f'Bars: {bars}, Wrappers: {wrappers}, Total: {total}')
        return total
    else:
        wrappers += 1
        bars -= 1
        total += 1
        print(f'Bars: {bars}, Wrappers: {wrappers}, Total: {total}')
        if wrappers == 4:
            bars += 1
            wrappers = 0
            # print(f'wrappers - 1: {wrappers}')
        return chocolateBars(bars, wrappers, total)
    # print(f'Bars: {bars}, Wrappers: {wrappers}, Total: {total}')    
print(chocolateBars(1000, 0, 0))
```
